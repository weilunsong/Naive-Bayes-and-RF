# Naive Bayes Classification

```python
from pyspark.mllib.classification import NaiveBayes, NaiveBayesModel
from pyspark.mllib.util import MLUtils
from pyspark.sql import SQLContext  
from pyspark.mllib.regression import LabeledPoint
```

```python
# Naive bayes classification for train data
sqlContest = SQLContext(sc)  
spark_df_train = sqlContest.createDataFrame(df_train)
traindata = spark_df_train.rdd.map(lambda x:LabeledPoint(x[-1],x[:-1])) # Convert pandas dataframe to pyspark dataframe
```

```python
model = NaiveBayes.train(spark_df_train.rdd.map(lambda x:LabeledPoint(x[-1],x[:-1])), 1.0) # Do naive bayes 
```

```python
predictionAndLabel = traindata.map(lambda p: (model.predict(p.features), p.label)) # Predict label
accuracy = 1.0 * predictionAndLabel.filter(lambda pl: pl[0] == pl[1]).count() / traindata.count() # Calculate accuracy
print('Train data accuracy {}'.format(accuracy))
```

```python
# Naive bayes classification for test data
spark_df_test = sqlContest.createDataFrame(df_test)
testdata = spark_df_test.rdd.map(lambda x:LabeledPoint(x[-1],x[:-1])) # Convert pandas dataframe to pyspark dataframe
```

```python
predictionAndLabel = testdata.map(lambda p: (model.predict(p.features), p.label))
accuracy = 1.0 * predictionAndLabel.filter(lambda pl: pl[0] == pl[1]).count() / testdata.count()
print('Test data accuracy {}'.format(accuracy))
```

```python
# Naive bayes classification for validation data
spark_df_vali = sqlContest.createDataFrame(df_vali)
validata = spark_df_vali.rdd.map(lambda x:LabeledPoint(x[-1],x[:-1])) # Convert pandas dataframe to pyspark dataframe
```

```python
predictionAndLabel = validata.map(lambda p: (model.predict(p.features), p.label))
accuracy = 1.0 * predictionAndLabel.filter(lambda pl: pl[0] == pl[1]).count() / validata.count()
print('Validation data accuracy {}'.format(accuracy))
```

# Random Forest

```python
from pyspark.mllib.tree import RandomForest, RandomForestModel
from pyspark.mllib.util import MLUtils
```

```python
model = RandomForest.trainClassifier(traindata, numClasses=4, categoricalFeaturesInfo={},
                                     numTrees=3, featureSubsetStrategy="auto",
                                     impurity='gini', maxDepth=4, maxBins=32)

model = RandomForest.trainClassifier(testdata, numClasses=4, categoricalFeaturesInfo={},
                                     numTrees=3, featureSubsetStrategy="auto",
                                     impurity='gini', maxDepth=4, maxBins=32)

model = RandomForest.trainClassifier(validata, numClasses=4, categoricalFeaturesInfo={},
                                     numTrees=3, featureSubsetStrategy="auto",
                                     impurity='gini', maxDepth=4, maxBins=32)
```

```python
# Random Forest for train data
predictions = model.predict(traindata.map(lambda x: x.features))
labelsAndPredictions = traindata.map(lambda lp: lp.label).zip(predictions)
trainErr = labelsAndPredictions.filter(
    lambda lp: lp[0] != lp[1]).count() / float(traindata.count())
print('Train Accuracay = ' + str(1-trainErr))
print('Learned classification forest model:')
print(model.toDebugString())
```

```python
# Random Forest for test data
predictions = model.predict(testdata.map(lambda x: x.features))
labelsAndPredictions = testdata.map(lambda lp: lp.label).zip(predictions)
testErr = labelsAndPredictions.filter(
    lambda lp: lp[0] != lp[1]).count() / float(testdata.count())
print('Test Accuracay = ' + str(1-testErr))
print('Learned classification forest model:')
print(model.toDebugString())
```

```python
# Random Forest for validation data
predictions = model.predict(validata.map(lambda x: x.features))
labelsAndPredictions = validata.map(lambda lp: lp.label).zip(predictions)
valiErr = labelsAndPredictions.filter(
    lambda lp: lp[0] != lp[1]).count() / float(validata.count())
print('Validation Accuracay = ' + str(1-valiErr))
print('Learned classification forest model:')
print(model.toDebugString())
```
